```python
fr = open('train_set.csv')
lines = fr.readlines()
lines[0]
```

```python
train_set = []
for line in lines:
    train_set.append( line.split(','))
train_set[1]
```

```python
#in train_set, the first is not used
train_set.pop(0)
#pre: the train_set
#post: all words make a table
def get_word_table( train_set ):
    word_table = {}
    for line in train_set:
        words = line[0].split(' ')
        for word in words:
            if word in word_table:
                pass
            else:
                pos = len( word_table)+1
                word_table[word] = pos
    word_table_1 = sorted( word_table.items() , key = lambda x : x[1] )
    return word_table_1
```

```python
word_table = get_word_table( train_set )
word_table
```
